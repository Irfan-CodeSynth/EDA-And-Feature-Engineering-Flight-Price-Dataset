## **Feature Eng. Processs**

```python
#  Feature Extractions for date column
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
df['Date'] = df['Date'].astype(int)
df['Month']= df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.info()
```

```python
#  Drop the date of journey column
df.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x: x.split(' ')[0])
```

```python
#  Feature Extraction for Arrival_Time columns to extract the hours and minuts

df['Arrival_Hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_Minutes'] = df['Arrival_Time'].str.split(':').str[1]

```

```python
df.head()
```

```python
#  Drop the arrival date column
df.drop('Arrival_date', axis=1, inplace=True)
```

```python
df.head()
```

```python
#  Drop the Arrival_Time column
df.drop('Arrival_Time', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Arrival_Hour'] = df['Arrival_Hour'].astype(int)
df['Arrival_Minutes']= df['Arrival_Minutes'].astype(int)
```

```python
df.info()
```

```python
df.head()
```

```python
#  Feature extraction of  Dep_Time
df['Dep_Hour'] = df['Dep_Time'].str.split(':').str[0]
df['Dep_Minutes'] = df['Dep_Time'].str.split(':').str[1]
```

```python
df.head()
```

```python
df['Dep_Minutes'] = df['Dep_Minutes'].astype(int)
df['Dep_Hour']= df['Dep_Hour'].astype(int)
```

```python
#  Drop the Dep_Time column
df.drop('Dep_Time', axis=1, inplace=True)
```

```python
#  Feature extraction for Duration columns
df['Duration_Hour'] =  df['Duration'].str.split(' ').str[0]
df['Duration_Minutes'] = df['Duration'].str.split(' ').str[1]
```
